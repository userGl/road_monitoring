```python
# ============================================
# –ù–ê–°–¢–†–û–ô–ö–ò –ü–†–û–ï–ö–¢–ê –ò –î–ê–¢–ê–°–ï–¢–ê
# ============================================

# –ò–º—è –ø—Ä–æ–µ–∫—Ç–∞ –∏ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞ (—Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–æ—Ö—Ä–∞–Ω—è—Ç—Å—è –≤ runs/_project/_name/)
_project = "runs/RoadDamage"
_name = "YOLOv8s_Baseline"

# –ü—É—Ç—å –∫ YAML —Ñ–∞–π–ª—É —Å –æ–ø–∏—Å–∞–Ω–∏–µ–º –¥–∞—Ç–∞—Å–µ—Ç–∞
# –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ dataset/road_damage.yaml –¥–ª—è —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–π —Å—Ç—Ä—É–∫—Ç—É—Ä—ã (train/val/test)
# –ò–ª–∏ dataset/rddJapanIndiaFiltered/rdd_JapanIndia.yaml –¥–ª—è —Å—Ç—Ä—É–∫—Ç—É—Ä—ã Japan/India
_data = "dataset/road_damage.yaml"

# ============================================
# –ü–ê–†–ê–ú–ï–¢–†–´ –¢–†–ï–ù–ò–†–û–í–ö–ò
# ============================================

# –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ø–æ—Ö (–ø–æ–ª–Ω—ã—Ö –ø—Ä–æ—Ö–æ–¥–æ–≤ –ø–æ –¥–∞—Ç–∞—Å–µ—Ç—É)
_epoch = 100

# Warmup —ç–ø–æ—Ö–∏ - –ø–æ—Å—Ç–µ–ø–µ–Ω–Ω–æ–µ —É–≤–µ–ª–∏—á–µ–Ω–∏–µ learning rate –≤ –Ω–∞—á–∞–ª–µ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–∫–∏
# –û–±—ã—á–Ω–æ 3-5% –æ—Ç –æ–±—â–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —ç–ø–æ—Ö
_warmup_epochs = int(_epoch * 0.05)  # 5% –æ—Ç _epoch = 5 —ç–ø–æ—Ö

# ============================================
# –ù–ê–°–¢–†–û–ô–ö–ê –£–°–¢–†–û–ô–°–¢–í–ê (GPU/CPU)
# ============================================

# –ê–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏–π –≤—ã–±–æ—Ä: CUDA –µ—Å–ª–∏ –¥–æ—Å—Ç—É–ø–Ω–∞, –∏–Ω–∞—á–µ CPU
_device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"–ò—Å–ø–æ–ª—å–∑—É–µ–º–æ–µ —É—Å—Ç—Ä–æ–π—Å—Ç–≤–æ: {_device}")

# ============================================
# –ó–ê–ì–†–£–ó–ö–ê –ú–û–î–ï–õ–ò
# ============================================

# –ó–∞–≥—Ä—É–∑–∫–∞ –ø—Ä–µ–¥–æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏ YOLOv8
# –î–æ—Å—Ç—É–ø–Ω—ã–µ –≤–∞—Ä–∏–∞–Ω—Ç—ã:
# - 'yolov8n.pt' - nano (—Å–∞–º–∞—è –º–∞–ª–µ–Ω—å–∫–∞—è, –±—ã—Å—Ç—Ä–∞—è)
# - 'yolov8s.pt' - small (–±–∞–ª–∞–Ω—Å —Å–∫–æ—Ä–æ—Å—Ç–∏ –∏ —Ç–æ—á–Ω–æ—Å—Ç–∏) ‚≠ê –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –∑–¥–µ—Å—å
# - 'yolov8m.pt' - medium
# - 'yolov8l.pt' - large
# - 'yolov8x.pt' - extra large (—Å–∞–º–∞—è —Ç–æ—á–Ω–∞—è, –º–µ–¥–ª–µ–Ω–Ω–∞—è)
model = YOLO('yolov8s.pt')

# ============================================
# –ó–ê–ü–£–°–ö –¢–†–ï–ù–ò–†–û–í–ö–ò
# ============================================

print(f"\n{'='*50}")
print(f"–ù–ê–ß–ê–õ–û –¢–†–ï–ù–ò–†–û–í–ö–ò")
print(f"{'='*50}")
print(f"–ü—Ä–æ–µ–∫—Ç: {_project}")
print(f"–≠–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç: {_name}")
print(f"–î–∞—Ç–∞—Å–µ—Ç: {_data}")
print(f"–≠–ø–æ—Ö: {_epoch}")
print(f"Warmup —ç–ø–æ—Ö: {_warmup_epochs}")
print(f"–£—Å—Ç—Ä–æ–π—Å—Ç–≤–æ: {_device}")
print(f"{'='*50}\n")

results = model.train(
    # –ü—É—Ç—å –∫ YAML —Ñ–∞–π–ª—É —Å –æ–ø–∏—Å–∞–Ω–∏–µ–º –¥–∞—Ç–∞—Å–µ—Ç–∞
    data=_data,
    
    # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ø–æ—Ö
    epochs=_epoch,
    
    # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ø–æ—Ö –¥–ª—è warmup (–ø–æ—Å—Ç–µ–ø–µ–Ω–Ω—ã–π —Ä–∞–∑–æ–≥—Ä–µ–≤ learning rate)
    warmup_epochs=_warmup_epochs,
    
    # –†–∞–∑–º–µ—Ä –±–∞—Ç—á–∞ (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –∑–∞ –æ–¥–Ω—É –∏—Ç–µ—Ä–∞—Ü–∏—é)
    # –ë–æ–ª—å—à–µ = –±—ã—Å—Ç—Ä–µ–µ, –Ω–æ —Ç—Ä–µ–±—É–µ—Ç –±–æ–ª—å—à–µ –ø–∞–º—è—Ç–∏ GPU
    # –î–ª—è RTX 4070 (8GB) –º–æ–∂–Ω–æ –ø–æ–ø—Ä–æ–±–æ–≤–∞—Ç—å 32, 16 –∏–ª–∏ –¥–∞–∂–µ 8
    batch=32,
    
    # –†–∞–∑–º–µ—Ä –≤—Ö–æ–¥–Ω–æ–≥–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è (–ø–∏–∫—Å–µ–ª–∏)
    # –ë–æ–ª—å—à–µ = —Ç–æ—á–Ω–µ–µ, –Ω–æ –º–µ–¥–ª–µ–Ω–Ω–µ–µ. –°—Ç–∞–Ω–¥–∞—Ä—Ç: 640
    imgsz=640,
    
    # –°–æ—Ö—Ä–∞–Ω—è—Ç—å —á–µ–∫–ø–æ–∏–Ω—Ç –∫–∞–∂–¥—ã–µ N —ç–ø–æ—Ö
    # –ü–æ–∑–≤–æ–ª—è–µ—Ç –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–∏—Ç—å —Ç—Ä–µ–Ω–∏—Ä–æ–≤–∫—É —Å –ª—é–±–æ–π —Å–æ—Ö—Ä–∞–Ω–µ–Ω–Ω–æ–π —ç–ø–æ—Ö–∏
    save_period=10,
    
    # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–æ—Ä–∫–µ—Ä–æ–≤ –¥–ª—è –∑–∞–≥—Ä—É–∑–∫–∏ –¥–∞–Ω–Ω—ã—Ö
    # –ë–æ–ª—å—à–µ = –±—ã—Å—Ç—Ä–µ–µ –∑–∞–≥—Ä—É–∑–∫–∞, –Ω–æ –±–æ–ª—å—à–µ –Ω–∞–≥—Ä—É–∑–∫–∞ –Ω–∞ CPU
    workers=4,
    
    # –£—Å—Ç—Ä–æ–π—Å—Ç–≤–æ –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏–π (cuda/cpu)
    device=_device,
    
    # –î–∏—Ä–µ–∫—Ç–æ—Ä–∏—è –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    project=_project,
    
    # –ò–º—è —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞ (–ø–æ–¥–¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—è –≤ project)
    name=_name,
    
    # Seed –¥–ª—è –≤–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    seed=1337,
    
    # Cosine learning rate scheduler
    # –ü–ª–∞–≤–Ω–æ —É–º–µ–Ω—å—à–∞–µ—Ç learning rate –ø–æ –∫–æ—Å–∏–Ω—É—Å–æ–∏–¥–µ
    cos_lr=True,
    
    # Mosaic augmentation (0.0 = –æ—Ç–∫–ª—é—á–µ–Ω–æ)
    # –°–º–µ—à–∏–≤–∞–Ω–∏–µ 4 –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –≤ –æ–¥–Ω–æ - –º–æ–∂–µ—Ç –ø–æ–º–æ—á—å, –Ω–æ –∏–Ω–æ–≥–¥–∞ –º–µ—à–∞–µ—Ç
    mosaic=0.0,
    
    # ============================================
    # –ü–ê–†–ê–ú–ï–¢–†–´ –õ–û–ì–ò–†–û–í–ê–ù–ò–Ø
    # ============================================
    
    # –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤ –∏ –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–π (True –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é)
    # –°–æ–∑–¥–∞–µ—Ç: results.png, confusion_matrix.png, F1_curve.png, PR_curve.png –∏ –¥—Ä.
    plots=True,
    
    # –ü–æ–¥—Ä–æ–±–Ω—ã–µ –ª–æ–≥–∏ –≤ –∫–æ–Ω—Å–æ–ª—å (True –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é)
    verbose=True,
    
    # –î–ª—è W&B –ª–æ–≥–∏—Ä–æ–≤–∞–Ω–∏—è (—Ä–∞—Å–∫–æ–º–º–µ–Ω—Ç–∏—Ä—É–π—Ç–µ, –µ—Å–ª–∏ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç–µ):
    # project="road-damage-detection",  # –ò–º—è –ø—Ä–æ–µ–∫—Ç–∞ –≤ W&B
    # name="YOLOv8s_Baseline",          # –ò–º—è —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞ –≤ W&B
    
    # –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –ø–æ–ª–µ–∑–Ω—ã–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã (–º–æ–∂–Ω–æ —Ä–∞—Å–∫–æ–º–º–µ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å):
    # patience=50,        # –†–∞–Ω–Ω—è—è –æ—Å—Ç–∞–Ω–æ–≤–∫–∞, –µ—Å–ª–∏ –Ω–µ—Ç —É–ª—É—á—à–µ–Ω–∏—è N —ç–ø–æ—Ö
    # lr0=0.01,          # –ù–∞—á–∞–ª—å–Ω—ã–π learning rate
    # lrf=0.01,          # –§–∏–Ω–∞–ª—å–Ω—ã–π learning rate (–∫–∞–∫ –¥–æ–ª—è –æ—Ç lr0)
    # momentum=0.937,    # Momentum –¥–ª—è –æ–ø—Ç–∏–º–∏–∑–∞—Ç–æ—Ä–∞
    # weight_decay=0.0005, # Weight decay (L2 —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è)
    # amp=True,          # Automatic Mixed Precision (—É—Å–∫–æ—Ä—è–µ—Ç —Ç—Ä–µ–Ω–∏—Ä–æ–≤–∫—É –Ω–∞ GPU)
)

print(f"\n{'='*50}")
print(f"–¢–†–ï–ù–ò–†–û–í–ö–ê –ó–ê–í–ï–†–®–ï–ù–ê!")
print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤: {_project}/{_name}/")
print(f"{'='*50}")
```
